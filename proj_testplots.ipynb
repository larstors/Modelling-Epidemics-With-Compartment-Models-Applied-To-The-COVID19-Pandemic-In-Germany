```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sol = np.genfromtxt('DiffEq.txt')
```

```python
lk=2
time=sol[:,0]
s=sol[:,1+lk]
i=sol[:,1+38+lk]
r=sol[:,1+2*38+lk]
d=sol[:,1+3*38+lk]
```

```python
plt.figure()
plt.plot(time,s,label='S')
plt.plot(time,i,label='I')
plt.plot(time,r,label='R')
plt.plot(time,d,label='D')
plt.ylim(0,1.2)
plt.legend()
plt.show()
```
